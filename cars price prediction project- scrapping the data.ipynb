```python
#Imports
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
driver =webdriver.Chrome(r"C:\Users\anand\Downloads\chromedriver_win32 (2)\chromedriver.exe")
driver.get("https://www.cars24.com/")
```

```python
Brand=[i.text for i in driver.find_elements(By.XPATH,"//h2[contains(@class,'_3FpCg')]")]
vriant=[i.text for i in driver.find_elements(By.XPATH,"//p[contains(@class,'cvakB')]")]
deatail=[i.text for i in driver.find_elements(By.XPATH,"//ul[contains(@class,'bVR0c')]")]
emi=[i.text for i in driver.find_elements(By.XPATH,"//div[contains(@class,'_2HFRN')]")]
price=[i.text for i in driver.find_elements(By.XPATH,"//div[contains(@class,'_7udZZ')]")]
down_payment=[i.text for i in driver.find_elements(By.XPATH,"//span[contains(@class,'_3X7mV _2H07I')]")]

data=list(zip(Brand,vriant,deatail,emi,price,down_payment))
df=pd.DataFrame(data,columns=['brand','model','details','emi','price','down payment'])
print("\ncars details")
df.head()
```
